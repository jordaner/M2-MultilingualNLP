```python
def get_sentences(sentences_path):
    with open(sentences_path, 'r') as s_file:
        return [" ".join(list(Text(line.strip()).words)) for line in s_file.readlines()]

def make_alignment_input(fr_sentences, en_sentences, alignment_path):
    with open(alignment_path, 'w') as output:
        for en_sentence, fr_sentence in zip(en_sentences, fr_sentences):
            output.write(f"{fr_sentence} ||| {en_sentence}\n")
```

# mBART Translation

```python
mbart_model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt").to(0)
mbart_tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
```

```python
def mbart_translate_sentence(model, tokenizer, sentence, src_lang="fr_XX", tgt_lang="en_XX"):
    """
    Translates a single sentence from the source language to the target language
    Using the model and tokenizer provided
    """
    tokenizer.src_lang = src_lang
    encoded_fr = tokenizer(sentence, return_tensors="pt").to(0)
    generated_tokens = model.generate(**encoded_fr, forced_bos_token_id=tokenizer.lang_code_to_id[tgt_lang])
    translated_sentence = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)[0]

    return translated_sentence
```

```python
df['mbart'] = df["reference"].progress_apply(lambda x: mbart_translate_sentence(mbart_model, mbart_tokenizer, x))
```

```python
import pandas as pd
import re

def find_target_in_translation(row, llm_name='mbart'):
    source = row['source']
    translation = row[llm_name]

    match = re.search(r'\[([^\]]+)\]', source)

    if match:
        target = match.group(1)

        if target in translation:
            return True
    return False
```

```python
result_table = df.groupby('category_major')['target_present'].mean().reset_index()

result_table = result_table.rename(columns={'target_present': 'accuracy'})

print("----- mbart accuracy by category -----")
print(result_table)
```

```python
df['target_present'] = df.apply(lambda row: find_target_in_translation(row, llm_name='mistral'), axis=1)

result_table = df.groupby('category_major')['target_present'].mean().reset_index()

result_table = result_table.rename(columns={'target_present': 'accuracy'})

print("----- mistral accuracy by category -----")
print(result_table)
```

# Misstral : Defining pipeline

```python
pipe = pipeline("text-generation",
                model="HuggingFaceH4/zephyr-7b-alpha",
                torch_dtype=torch.bfloat16,
                device_map="auto")
```

```python
def mistral_translate(sentence):
  messages = [{"role": "user", "content": f"traduit en anglais: {sentence}"}]
  prompt = pipe.tokenizer.apply_chat_template(messages,
                                              tokenize=False,
                                              add_generation_prompt=True)

  outputs = pipe(prompt,
                  max_new_tokens=256,
                  do_sample=True,
                  temperature=0.7,
                  top_k=50,
                  top_p=0.95)

  return outputs[0]['generated_text'].split('<|assistant|>\n')[1]
```

```python
df['mistral'] = df["reference"].progress_apply(lambda x: mistral_translate(x))
```
